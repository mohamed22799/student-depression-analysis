# 1) Distribution of Depression Among Genders

# 2) Effect of Academic Pressure on Depression

```python
# Convert Depression column from 1/0 to Yes/No
df['Depression']=df['Depression'].replace({
    0: 'No',
    1:'Yes'
})
```

### Interpretation:

#### For students with Depression:

- the interquartile range (IQR) (spread of the box) indicates that students with higher depression levels are mostly between 3:5.
- there are a noticeable concetration of data in the upper end, indicating that the depression levels for these students tend to report higher academic pressure.
- a few outliers exist, but most values cluster to upper end.

#### For students without Depression:
- the medien value for the academic pressure is lower(around 2)
- the IQR falls between 1:3
- there are a few outliers at higher end (up to 5), but the majority of these students to to report lower academic pressure.
  

# 3) Correlation Between CGPA and Sleep Duration

```python
df['Sleep Duration'].unique()
```

```python
# Replace ranges with average numeric values
df['Sleep Duration']=df['Sleep Duration'].replace({
    '5-6 hours': 5.5,
    'Less than 5 hours': 4,
    '7-8 hours':7.5,
    'More than 8 hours':9,
    'Others': None
})

# Convert column to numeric and drop rows with 'Others' or missing values
df['Sleep Duration']=pd.to_numeric(df['Sleep Duration'],errors='coerce')
```

```python
print(df['Sleep Duration'].unique())
```

```python
df['Sleep Duration'].value_counts()
```

```python
# Scatter Plot
sns.lmplot(x='CGPA',y='Sleep Duration', data=df)
plt.title('CGPA VS. Sleep Duration', fontsize=20)
plt.xlabel('CGPA', fontsize=12)
plt.ylabel('Sleep Duration', fontsize=12)
plt.show()
```

```python
correlation = df['CGPA'].corr(df['Sleep Duration'])
print('Correlation between CGPA and Sleep Duration',correlation)
```

### Interpretation: 

- The line of best fit is almost flat, which suggests there is no strong correlation between CGPA and Sleep Duration.
- Whether students sleep less (4-5 hours) or more (7-9 hours), their CGPA does not show significant variation.
- The plot indicates no significant relationship between CGPA and Sleep Duration.
- This means that, for this dataset, the amount of sleep a student gets does not strongly influence their academic performance (CGPA).

# 6) Average Work/Study Hours Based on Gender

```python
df['Gender'].unique()
```

# 7) Average Work/Study Hours by Depression Level

```python
avg_hours_depression = df.groupby('Depression')['Work/Study Hours'].mean().reset_index()
print(avg_hours_depression)

sns.barplot(x='Depression', y='Work/Study Hours', data=avg_hours_depression, hue='Depression', palette='Set2', legend=False)
plt.title('Average Work/Study Hours by Depression Level', fontsize=16)
plt.xlabel('Depression Level', fontsize=12)
plt.ylabel('Average Work/Study Hours', fontsize=12)
for container in plt.gca().containers:
    plt.bar_label(container, fmt='%.2f', label_type='edge', fontsize=10, padding=1)
plt.show()
```

### Interpretation:
- The graph shows a relationship between average work/study hours and depression levels.
- Students experiencing depression tend to have **higher average work/study hours** (approximately 7.8 hours) compared to students who are not depressed, whose average is **around 6.2 hours**.
- This indicates a possible correlation where an **increase in work/study hours** might contribute to higher levels of stress, which could be linked to depression.
- **Overall**: Managing work/study hours effectively might play a role in reducing stress and mitigating depression.

# 10) Impact of Study Satisfaction on Depression

```python
sns.boxplot(x='Depression', y='Study Satisfaction', hue='Depression', data=df, palette='coolwarm')
plt.title('Study Satisfaction vs Depression', fontsize=20)
plt.xlabel('Depression (0 = No, 1 = Yes)',fontsize=12)
plt.ylabel('Study Satisfaction',fontsize=12)
plt.show()
```

### Interpretation:
- The box plot compares study satisfaction levels between individuals who experience depression (coded as 1 or "Yes") and those who do not (coded as 0 or "No").  Here's an interpretation:

*   **Overall Distribution:** The boxes for both groups (depressed and non-depressed) overlap significantly, suggesting that there isn't a large difference in the *median* study satisfaction between the two groups.

*   **Median:** The horizontal line within each box represents the median. It appears that the medians for both groups are relatively close, around 3. This indicates that the "typical" study satisfaction score is similar regardless of depression status.

*   **Interquartile Range (IQR):** The box itself represents the interquartile range (IQR), which contains the middle 50% of the data. The similar size of the boxes suggests that the spread or variability of study satisfaction scores is comparable in both groups.

*   **Whiskers and Outliers:** The lines extending from the boxes (whiskers) show the range of the data, excluding outliers. Both groups have similar whisker lengths, indicating a similar range of satisfaction scores. There appear to be no extreme outliers plotted individually as points outside the whiskers in either group.

*   **Conclusion:** Based on this box plot, there is no strong visual evidence to suggest a substantial difference in study satisfaction levels between individuals who experience depression and those who do not. While there might be subtle differences or other factors at play, the central tendency (median) and spread (IQR) of study satisfaction appear quite similar across both groups. Statistical tests (like a t-test or Mann-Whitney U test) would be needed to formally confirm whether any observed differences are statistically significant.
